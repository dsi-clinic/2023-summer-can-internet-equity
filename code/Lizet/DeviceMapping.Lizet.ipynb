```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns
import seaborn.objects as so
```

```python
Casas_device = pd.read_csv('/srv/data/my_shared_data_folder/internet-equity/device_data_oct21//nm-mngd-20210511-9d9824d4.csv')
Casas_device
```

```python
Casas_device_At = Casas_device.loc[Casas_device['Atlanta'] == 'Atlanta']
Casas_device_At
```

```python
Casas_device_At.ping_latency.unique()
```

```python
Casas_device_avg = Casas_device_At.loc[Casas_device_At['avg'] == 'avg']
Casas_device_avg
```

```python
Casas_device_min = Casas_device_At.loc[Casas_device_At['avg'] == 'min']
Casas_device_min
```

```python
Casas_device_max = Casas_device_At.loc[Casas_device_At['avg'] == 'max']
Casas_device_max
```

```python
Casas_device_mdev = Casas_device_At.loc[Casas_device_At['avg'] == 'mdev']
Casas_device_mdev
```

```python
fig,ax =  plt.subplots( 2, 2,
                       figsize = ( 10, 8))
  
sns.lineplot( x = "2021-10-27 02:51:03", y = "37.803", 
             color = 'r', data = Casas_device_avg, 
             ax = ax[0][0])
  
ax[0][0].tick_params(labelrotation = 25)
sns.lineplot( x = "2021-10-27 02:51:03", y = "37.803", 
             color = 'g', data = Casas_device_min,
             ax = ax[0][1])
  
ax[0][1].tick_params(labelrotation = 25)
sns.lineplot(x = "2021-10-27 02:51:03", y = "37.803", 
             color = 'b', data = Casas_device_max,
             ax = ax[1][0])
  
ax[1][0].tick_params(labelrotation = 25)
  
sns.lineplot(x = "2021-10-27 02:51:03", y = "37.803", 
             color = 'y', data = Casas_device_mdev, 
             ax = ax[1][1])
  
ax[1][1].tick_params(labelrotation = 25)
fig.tight_layout(pad = 1.2)
```
