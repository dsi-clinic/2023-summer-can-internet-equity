```python
import pandas as pd
```

```python
device_5atl = device_5.loc[device_5['Atlanta'] == 'Atlanta']
device_5atl
```

```python
device_5mdev = device_5.loc[device_5['avg'] == 'mdev']
device_5mdev
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig,ax =  plt.subplots( 2, 2,
                       figsize = ( 10, 8))
  
sns.lineplot( x = "2021-10-18 13:01:01", y = "26.59", 
             color = 'r', data = device_5avg, 
             ax = ax[0][0])
  
ax[0][0].tick_params(labelrotation = 25)
sns.lineplot( x = "2021-10-18 13:01:01", y = "26.59", 
             color = 'g', data = device_5min,
             ax = ax[0][1])
  
ax[0][1].tick_params(labelrotation = 25)
sns.lineplot(x = "2021-10-18 13:01:01", y = "26.59", 
             color = 'b', data = device_5max,
             ax = ax[1][0])
  
ax[1][0].tick_params(labelrotation = 25)
  
sns.lineplot(x = "2021-10-18 13:01:01", y = "26.59", 
             color = 'y', data = device_5mdev, 
             ax = ax[1][1])
  
ax[1][1].tick_params(labelrotation = 25)
fig.tight_layout(pad = 1.2)
```
