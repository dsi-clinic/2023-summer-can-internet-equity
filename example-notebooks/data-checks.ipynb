```python
import pandas as pd
import os

data_path = "../data/netrics/2022-07-01-2022-08-01"
```

```python
speedtests = pd.read_csv(os.path.join(data_path, "speedtest.csv"))
```

```python
print(speedtests.shape)
speedtests.columns
```

```python
print(f"Date range: {speedtests.Time.min()} - {speedtests.Time.max()}")
```

```python
pd.options.display.max_rows = 300
print(speedtests.Direction.unique())
print(speedtests.Tool.unique())
print("Speed summaries\n", speedtests.Speed.describe())
print("Number of devices: ", speedtests.ID.nunique())
speedtests.groupby(["ID", "Tool"]).size().reset_index()
```
